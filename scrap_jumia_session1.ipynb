# Formation Machine Learning: Collecte de données par le scrapping sur Jumia

```python
#import requests
import csv
import re
import codecs
#import ssl
#import time
from bs4 import BeautifulSoup as soup
from urllib.request import Request, urlopen
#from csv import writer
```

```python
#creation fichier csv et ouvrir
f= codecs.open('jumia.csv','wb',encoding= 'utf-8') 
writer = csv.writer(f)
# ecrire dans le csv des entretes à partir des infos à collecter à partir du site
writer.writerow(["description","remise", "prix","ancien_prix"])
for i in range(1,51): 
    print("Numéro de la page à scrapper:",i)
    #req = Request('https://www.jumia.com.tn/telephone-tablettes/?page={accolade à mettre pour s'incrementer (nombre de pages)}'.format(i), headers={'User-Agent': 'Mozilla/5.0'})
    req = Request('https://www.jumia.com.tn/electronique/?page={}#catalog-listing'.format(i), headers={'User-Agent': 'Mozilla/5.0'})   
    webpage = urlopen(req).read()
    page_soup=soup(webpage,'html.parser')
    page_soup.text
    #print(page_soup)
    posts=page_soup.findAll('div',class_='info')
    print("Le nombre d'item est:",len(posts))
    #print(posts)
    print('******************************************')
    for post in posts:     
        print(post)
        description=post.findAll('h3', class_='name')
        if not description:
            description2 = ""
        else:
            description1=description[0].text
            description2 = re.sub('\s+',' ',description1)
            print(description2)
        
        remise=post.findAll('div', class_='bdg _dsct _sm')
        if not remise:
            remise2=""
        else:
            remise1=remise[0].text
            remise2 = re.sub('\s+',' ',remise1)
            print(remise2)
        
        price=post.findAll('div', class_='prc')
        if not price:
            price2=""
        else:
            price1=price[0].text
            price2 = re.sub('\s+',' ',price1)
            print(price2)        
        old_price=post.findAll('div', class_='old')
        if not old_price:
            old_price=""
        else:
            old_price1=old_price[0].text
            old_price2 = re.sub('\s+',' ',old_price1)
            print(old_price2)
        if price2=="" and description2 == "":
            print("bonjour")
        else:
            writer.writerow([description2, remise2, price2,old_price2])
        print("ok")
    print('*****************************************')
    print('*****************************************')
    print('*****************************************')
    print('*****************************************')
f.close()
```
